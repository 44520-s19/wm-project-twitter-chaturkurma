## Storing the Retrieved information or Data in Pickle File

&nbsp; &nbsp;&nbsp; In this Step i have collected the Data or Information from the Twitter and Store that data in the pickle file.

```python
import tweepy
import pickle
from twitter_keys import consumer_key, consumer_secret, access_token, access_secret

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)
api = tweepy.API(auth)

obama_tweets = api.user_timeline('BarackObama')
with open('obama.pkl', 'wb') as f:
    pickle.dump(obama_tweets, f)

praba_tweets = api.user_timeline('ActorPRABHA')
with open('praba.pkl', 'wb') as f:
    pickle.dump(praba_tweets, f)
    
katy_tweets = api.user_timeline('katyperry')
with open('katy.pkl', 'wb') as f:
    pickle.dump(katy_tweets, f)
```

# Analysis Steps

# Lexical Diversity

 &nbsp; &nbsp;&nbsp; Lexical Diversity is the average number of times a word is repeated in the text. It is simply the ratio of the total number of words in the text to the number of unique words in the text.

  I got three different lexical diversity's for each tweet they are:
  
 - Lexical diversity for obama tweet is 0.85
 - Lexical diversity for praba tweet is 0.91666
 - Lexical diversity for katy tweet is 0.888888

 Among the three tweets the lexical diversity is more for the praba tweet because there are more number of repeated words in that text when compared to remaining other two tweets.


```python
def lexical_diversity(text):
    tokens = text.lower().split()
    return len(set(tokens))/len(tokens)

print('lexical diversity for obama tweet:', lexical_diversity(obama_tweets[0].text))
print('lexical diversity for praba tweet:', lexical_diversity(praba_tweets[0].text))
print('lexical diversity for katy tweet:', lexical_diversity(katy_tweets[0].text))
```

# Commonly used words

&nbsp;&nbsp;&nbsp; It shows the number of times each word is used or repeated in the text . It results in how many times the word is used in text. Using counter we can find the total count of word in text.

```python
paragraph = "The Beginning is a 2015 Indian epic action film directed by S. S. Rajamouli. The film was produced by Shobu Yarlagadda and Prasad Devineni and was shot in both Telugu and Tamil. This film was also dubbed into Malayalam and Hindi. The film stars Prabhas, Rana Daggubati, Anushka Shetty, and Tamannaah in the lead roles, with Ramya Krishna, Sathyaraj, and Nassar in supporting roles. The first of two cinematic parts, the film follows Sivudu, an adventurous young man who helps his love, a rebellious warrior intending to rescue the former queen of Mahishmati. In the process, he learns of his true identity as the heir to the throne of Mahishmati, the son of Amarendra Baahubali, whose tale is narrated to him by Kattappa, a loyal warrior. When asked about Amarendras whereabouts, Kattappa reveals that he was the one who killed him. The reason behind this is explained in the follow-up Baahubali 2: The Conclusion."
from collections import Counter
words = paragraph.lower().split()
word_count = Counter(words)
print(word_count)
```

## Visualization of Data
&nbsp;&nbsp;&nbsp; It is the Pictographic representation of the Histogram of different words occured to the word frequency. 

```python
import matplotlib.pyplot as plt
NUM_VALUES = 10
heights = list(sorted(word_count.values(), reverse=True))
print(heights)
plt.bar(range(NUM_VALUES),heights[:NUM_VALUES])
plt.title('word frenquencies')
plt.ylabel('# Occurences')
plt.xlabel('Different words')
keys = list(word_count.keys())
keys.sort(key= lambda x: word_count[x], reverse=True)
plt.xticks(range(NUM_VALUES), keys[:NUM_VALUES])
print(keys)
plt.show()
```

# Generating wordcloud
 &nbsp;&nbsp;&nbsp; Word Cloud is a data visualization technique used for representing text data in which the size of each word indicates its frequency or importance. Significant textual data points can be highlighted using a word cloud. 

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

paragraph = "The Beginning is a 2015 Indian epic action film directed by S. S. Rajamouli. The film was produced by Shobu Yarlagadda and Prasad Devineni and was shot in both Telugu and Tamil. This film was also dubbed into Malayalam and Hindi. The film stars Prabhas, Rana Daggubati, Anushka Shetty, and Tamannaah in the lead roles, with Ramya Krishna, Sathyaraj, and Nassar in supporting roles. The first of two cinematic parts, the film follows Sivudu, an adventurous young man who helps his love, a rebellious warrior intending to rescue the former queen of Mahishmati. In the process, he learns of his true identity as the heir to the throne of Mahishmati, the son of Amarendra Baahubali, whose tale is narrated to him by Kattappa, a loyal warrior. When asked about Amarendras whereabouts, Kattappa reveals that he was the one who killed him. The reason behind this is explained in the follow-up Baahubali 2: The Conclusion."
from collections import Counter
words = paragraph.lower().split()
word_count = Counter(words)
print(word_count)

wordcloud = WordCloud(width = 1000, height = 1000, min_font_size = 10).generate(paragraph)
plt.figure(figsize = (10, 10))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

# Sentiment Analysis
&nbsp;&nbsp;&nbsp; Sentiment Analysis is a process to determine a tweet is Positive or Negative or Neutral.

```python
import pickle 
from nltk_helpers import get_sentiments 
    
print(get_sentiments(obama_tweets[0].text))
print(get_sentiments(praba_tweets[0].text))
print(get_sentiments(katy_tweets[0].text))

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
# build our figure
fig = plt.figure()
# build the axes
ax = fig.add_subplot(111, projection='3d')

import pickle
from nltk_helpers import split_sentiments

ac1X, ac1Y, ac1Z = split_sentiments([get_sentiments(tweet.text) for tweet in obama_tweets])
ax.scatter(ac1X, ac1Y, ac1Z, color = 'g', marker = 'd')

ac2X, ac2Y, ac2Z = split_sentiments([get_sentiments(tweet.text) for tweet in praba_tweets])
ax.scatter(ac2X, ac2Y, ac2Z, color = 'b', marker = 'o')

ac3X, ac3Y, ac3Z = split_sentiments([get_sentiments(tweet.text) for tweet in katy_tweets])
ax.scatter(ac3X, ac3Y, ac3Z, color = 'r', marker = 'p')


ax.set_xlabel('Negative')
ax.set_ylabel('Neutral')
ax.set_zlabel('Positive')

plt.show()

```

```python
plt.scatter(ac1X, ac1Y, color = 'g', marker = 'd', label = 'Obama')
plt.scatter(ac2X, ac2Y, color = 'b', marker = 'o', label = 'Prabhas')
plt.scatter(ac3X, ac3Y, color = 'r', marker = 'p', label = 'Katy')
plt.xlabel('Negative')
plt.ylabel('Positive')
plt.title('Positive vs Negative Sentiments')
plt.legend(loc = 'upper right', shadow = True)
plt.show()
```
